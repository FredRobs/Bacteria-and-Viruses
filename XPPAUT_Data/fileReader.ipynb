## Modulation chemostat two parameter bifurcation m and omega

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df1=pd.read_csv('data1.csv', names=['omega', 'm'])
df2=pd.read_csv('data2.csv', names=['omega', 'm'])
df3=pd.read_csv('data3.csv', names=['omega', 'm'])
df4=pd.read_csv('data4.csv', names=['omega', 'm'])

plt.figure(figsize=(7,7))

torusBranch1=np.linspace(0.0033, 0.0042, 22)
torusBranch2=np.linspace(0.0037, 0.00552, 279)
torusBranch3=np.linspace(0.005003, 0.00555, 15)
torusBranch4=np.linspace(0.00552, 0.0040, 279)

coeffBranch1=np.polyfit(df1['omega'], df1['m'], 1)
coeffBranch2=np.polyfit(df4['omega'], df4['m'], 5)
a3, b3=np.polyfit(df3['omega'], df3['m'], 1)
a4, b4=np.polyfit(df2['omega'], df2['m'], 1)

torusSolid = torusBranch2[torusBranch2 <= 0.005]
torusDashed = torusBranch2[torusBranch2 >= 0.005]

p1 = np.poly1d(coeffBranch1)
p2 = np.poly1d(coeffBranch2)

plt.plot(torusBranch1-0.000014, p1(torusBranch1),color='black',linewidth=1)
plt.plot(torusSolid, p2(torusSolid),color='black',linewidth=1)
plt.plot(torusBranch3, a3*torusBranch3+b3,color='black',linewidth=1)
plt.plot(torusBranch4, a4*torusBranch4+b4,'--',color='black',linewidth=1)
plt.plot(torusDashed, p2(torusDashed),'--',color='black',linewidth=1)

#plt.scatter(df1['omega'], df1['m'], color='black',s=2.0,alpha=0.5)
#plt.scatter(df2['omega'], df2['m'], color='black',s=2.0,)
#plt.scatter(df3['omega'], df3['m'], color='black', s=2.0,alpha=0.5)
#plt.scatter(df2['omega'], df2['m'], color='blue', s=13.0,alpha=0.5)

df5=pd.read_csv('data5.csv', names=['omega', 'm'])
df6=pd.read_csv('data6.csv', names=['omega', 'm'])

homoclinicBranch1=np.linspace(0.0037, 0.00552, 279)
coeffBranch3=np.polyfit(df6['omega'], df6['m'], 3)
p3 = np.poly1d(coeffBranch3)

#plt.scatter(df6['omega'], df6['m'], color='blue', s=13.0,alpha=0.5)
#plt.scatter(df6['omega'], df6['m'], color='blue', s=13.0,alpha=0.5)

plt.fill_betweenx(df5['m'], 0.0, df5['omega'], color='lightsalmon', alpha=0.15)
plt.fill_betweenx(p3(homoclinicBranch1), 0.01, homoclinicBranch1, color='lightsalmon', alpha=0.15)
plt.plot(df5['omega'], df5['m'], color='black',linewidth=1)
plt.plot(homoclinicBranch1, p3(homoclinicBranch1),color='black',linewidth=1)
plt.xlabel("$\Omega$", fontsize=18)
plt.ylabel("M", fontsize=18)
plt.xlim(0.003,0.0055)
plt.ylim(0,0.2)

plt.text(0.0036, 0.16, '$SN$', fontsize=12, fontstyle='italic')
plt.text(0.0045, 0.033, '$SN$', fontsize=12, fontstyle='italic')
plt.text(0.00502, 0.1, '$T$', fontsize=12, fontstyle='italic')
plt.text(0.00398, 0.13, '$H$', fontsize=12, fontstyle='italic')
plt.text(0.00485, 0.13, '$H$', fontsize=12, fontstyle='italic')
plt.text(0.00543, 0.1, '$SSN$', fontsize=12, fontstyle='italic')
plt.scatter(0.005, 0.0695,marker='o',color='black')
plt.text(0.00504, 0.0665, '$T-B$', fontsize=12, fontstyle='italic')
plt.scatter(0.00551, 0.0642,marker='o',color='black')

plt.scatter(0.0039, 0.1,marker='o',color='black', s=5)
plt.text(0.00386, 0.105, '$1$', fontsize=12, fontstyle='italic')

plt.scatter(0.0040, 0.1,marker='o',color='black', s=5)
plt.text(0.00398, 0.105, '$2$', fontsize=12, fontstyle='italic')

plt.scatter(0.0048, 0.1,marker='o',color='black', s=5)
plt.text(0.00478, 0.105, '$3$', fontsize=12, fontstyle='italic')

plt.scatter(0.0049, 0.1,marker='o',color='black', s=5)
plt.text(0.00489, 0.105, '$4$', fontsize=12, fontstyle='italic')

plt.text(0.00555, 0.0632, '$T-B$', fontsize=12, fontstyle='italic')
plt.xlabel("$\Omega$", fontsize=18)
plt.ylabel("$m$", fontsize=18)
plt.xlim(0.0032,0.0058)
plt.ylim(0,0.2)
```

## Two parameter bifurcation Cascaded

```python
df1=pd.read_csv('PDBranch1.csv', names=['sigma_2', 'sigma_1'])
df2=pd.read_csv('extinctionPD.csv', names=['sigma_2', 'sigma_1'])
df4=pd.read_csv('hopfPD.csv', names=['sigma_2', 'sigma_1'])
df5=pd.read_csv('hopf02PD.csv', names=['sigma_2', 'sigma_1'])
df6=pd.read_csv('Cascaded01.csv', names=['sigma_2', 'sigma_1'])
df7=pd.read_csv('Cascaded02.csv', names=['sigma_2', 'sigma_1'])
df8=pd.read_csv('Cascaded04.csv', names=['sigma_2', 'sigma_1'])
df9=pd.read_csv('Cascaded05.csv', names=['sigma_2', 'sigma_1'])
df10=pd.read_csv('Cascaded06.csv', names=['sigma_2', 'sigma_1'])
plt.figure(figsize=(7,7))

CascadedBranch1=np.linspace(0.0, 14, 279)
coeffBranch1=np.polyfit(df6['sigma_2'], df6['sigma_1'], 3)
p1 = np.poly1d(coeffBranch1)
CascadedBranch2=np.linspace(0.0, 14, 279)
coeffBranch2=np.polyfit(df7['sigma_2'], df7['sigma_1'], 2)
p2 = np.poly1d(coeffBranch2)
CascadedBranch3=np.linspace(0.0,14, 279)
coeffBranch3=np.polyfit(df8['sigma_2'], df8['sigma_1'], 2)
p3 = np.poly1d(coeffBranch3)
CascadedBranch4=np.linspace(0.0,14, 279)
coeffBranch4=np.polyfit(df9['sigma_2'], df9['sigma_1'], 2)
p4 = np.poly1d(coeffBranch4)
CascadedBranch5=np.linspace(1.67,14, 279)
coeffBranch5=np.polyfit(df10['sigma_2'], df10['sigma_1'], 2)
p5 = np.poly1d(coeffBranch5)

plt.plot(df1['sigma_2'], df1['sigma_1'], color='black')
plt.plot(df2['sigma_2'], df2['sigma_1'], color='black')
plt.plot(df4['sigma_2'], df4['sigma_1'], color='black')
plt.plot(df5['sigma_2'], df5['sigma_1'], color='black')
plt.plot(CascadedBranch1, p1(CascadedBranch1),color='black')
plt.plot(CascadedBranch2, p2(CascadedBranch2),color='black')
plt.plot(CascadedBranch3, p3(CascadedBranch3),'--',color='black')
plt.plot(CascadedBranch4, p4(CascadedBranch4),color='black')
plt.plot(CascadedBranch5, p5(CascadedBranch5),color='black')

plt.text(5.5, 16, '$SN^2$', fontsize=12, fontstyle='italic')
plt.text(5.9, 15.14, '$PD^1_2$', fontsize=12, fontstyle='italic')
plt.text(6.1, 13.3, '$PD^1_1$', fontsize=12, fontstyle='italic')
plt.text(1, 15, '$Chaos^2$', fontsize=12, fontstyle='italic')
plt.text(7.8, 13.3, '$G$', fontsize=12, fontstyle='italic')
plt.text(8, 19, '$Chaos^1$', fontsize=12, fontstyle='italic')
plt.text(4.2, 18, '$PD^2_1$', fontsize=12, fontstyle='italic')

clb = np.minimum(p2(CascadedBranch2), p4(CascadedBranch4))

# Fills the area between the p4 curve and the horizontal line y=12
plt.fill_between(CascadedBranch4, clb, 20, color='lightsalmon', alpha=0.15)
plt.fill_between(CascadedBranch2, clb, 20, color='lightsalmon', alpha=0.15)

plt.xlabel("$\sigma_2$", fontsize=15)
plt.ylabel("$\sigma_1$", fontsize=15)

plt.xlim(0,13)
plt.ylim(6.5,20)
```
