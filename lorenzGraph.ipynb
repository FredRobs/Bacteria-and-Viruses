```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
s=10
r=28
b=8/3
t_start=0
t_end=100

x_init=0
y_init=1
z_init=0
initialState=[x_init,y_init,z_init]

def odeModel(t,y):
    x,y,z=y
    x_dot = s * (y - x)
    y_dot = r * x - y - x * z
    z_dot = x * y - b * z

    return [x_dot,y_dot,z_dot]
```

```python
solution = solve_ivp(odeModel,[t_start,t_end],initialState,method='RK45',t_eval=np.linspace(t_start, t_end, 12000))
```

```python
x_sol=solution.y[0]
y_sol=solution.y[1]
z_sol=solution.y[2]
plt.plot(x_sol,z_sol,color='black',linewidth=0.5)
plt.ylabel("Z")
plt.xlabel("X")
```
