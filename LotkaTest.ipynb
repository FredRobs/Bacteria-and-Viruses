```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
#Constants
a=2
b=1.1
gamma=1
delta=0.9
Lambda=0.1
k=0.1
t_start=0
t_end=20
t_eval=np.linspace(t_start, t_end, 10000)
```

```python
#Initial state of system
X_init=1
Y_init=0.5
initialState=[X_init,Y_init]
```

```python
#ODE model for lotka
def odeModel(t,y):
    X,Y=y
    X_dot=a*X - b*X*Y
    Y_dot=-gamma*Y + delta*X*Y

    return [X_dot,Y_dot]

#ODE model for lotka w fluctuations
def odeModelF(t,y):
    X,Y=y
    X_dot=a*X - Lambda*(X)**2 - b*X*Y
    Y_dot=-gamma*Y - k*(Y)**2 + delta*X*Y

    return [X_dot,Y_dot]


solution = solve_ivp(odeModel,[t_start,t_end],initialState,method='DOP853',t_eval=t_eval)
solutionF = solve_ivp(odeModelF,[t_start,t_end],initialState,method='DOP853',t_eval=t_eval)
```

```python

X_sol=solution.y[0]
Y_sol=solution.y[1]
X_solF=solutionF.y[0]
Y_solF=solutionF.y[1]

plt.plot(X_sol,Y_sol)
plt.ylabel("Y")
plt.xlabel("X")
plt.grid()
plt.show()
```
