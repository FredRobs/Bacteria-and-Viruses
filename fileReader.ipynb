```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df1=pd.read_csv('data1.csv', names=['omega', 'm'])
df2=pd.read_csv('data2.csv', names=['omega', 'm'])
df3=pd.read_csv('data3.csv', names=['omega', 'm'])
df4=pd.read_csv('data4.csv', names=['omega', 'm'])

plt.figure(figsize=(9,9))

torusBranch1=np.linspace(0.0033, 0.0042, 22)
torusBranch2=np.linspace(0.0037, 0.00552, 279)
torusBranch3=np.linspace(0.005003, 0.00555, 15)
torusBranch4=np.linspace(0.00552, 0.0040, 279)

coeffBranch1=np.polyfit(df1['omega'], df1['m'], 1)
coeffBranch2=np.polyfit(df4['omega'], df4['m'], 5)
a3, b3=np.polyfit(df3['omega'], df3['m'], 1)
a4, b4=np.polyfit(df2['omega'], df2['m'], 1)

torusSolid = torusBranch2[torusBranch2 <= 0.005]
torusDashed = torusBranch2[torusBranch2 >= 0.005]

p1 = np.poly1d(coeffBranch1)
p2 = np.poly1d(coeffBranch2)

plt.plot(torusBranch1, p1(torusBranch1),color='black',linewidth=1)
plt.plot(torusSolid, p2(torusSolid),color='black',linewidth=1)
plt.plot(torusDashed, p2(torusDashed),'--',color='black',linewidth=1)
plt.plot(torusBranch3, a3*torusBranch3+b3,color='black',linewidth=1)
plt.plot(torusBranch4, a4*torusBranch4+b4,'--',color='black',linewidth=1)

#plt.scatter(df1['omega'], df1['m'], color='black',s=2.0,alpha=0.5)
#plt.scatter(df2['omega'], df2['m'], color='black',s=2.0,)
#plt.scatter(df3['omega'], df3['m'], color='black', s=2.0,alpha=0.5)
#plt.scatter(df2['omega'], df2['m'], color='blue', s=13.0,alpha=0.5)
plt.text(0.0036, 0.16, '$SN$', fontsize=10, fontstyle='italic')
plt.text(0.0045, 0.033, '$SN$', fontsize=10, fontstyle='italic')
plt.text(0.00502, 0.1, '$T$', fontsize=10, fontstyle='italic')
plt.text(0.00543, 0.1, '$SSN$', fontsize=10, fontstyle='italic')
plt.scatter(0.005, 0.0695,marker='o',color='black')
plt.text(0.00504, 0.0665, '$T-B$', fontsize=10, fontstyle='italic')
plt.scatter(0.00551, 0.0642,marker='o',color='black')
plt.text(0.00555, 0.0632, '$T-B$', fontsize=10, fontstyle='italic')
plt.xlabel("$\Omega$")
plt.ylabel("M")
plt.xlim(0.003,0.0058)
plt.ylim(0,0.2)
```
